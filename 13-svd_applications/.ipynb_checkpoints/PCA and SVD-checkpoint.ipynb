```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt


import seaborn as sns
sns.set(font_scale=2)
plt.style.use('seaborn-whitegrid')
%matplotlib inline

np.random.seed(8885)

from sklearn.decomposition import PCA
```

## Take some random data in 2d space

```python
Xhat = np.dot(np.random.rand(2, 2), np.random.randn(2, 200)).T
#Xhat = np.random.randn(2, 200).T
print(Xhat.shape)

M = Xhat.shape[0]
N = Xhat.shape[1]


plt.scatter(Xhat[:, 0], Xhat[:, 1]);
plt.axis('equal');
plt.xlabel('x1')
plt.ylabel('x2')
```

## "Remove" the mean

```python
X = Xhat - Xhat.mean(axis=0)
plt.scatter(X[:, 0], X[:, 1]);
plt.axis('equal');
plt.xlabel('x1-mean1')
plt.ylabel('x2-mean2')
```

## Assemble the "data" matrix

```python
Y = X/np.sqrt(M-1)
plt.scatter(Y[:, 0], Y[:, 1]);
plt.axis('equal');
plt.xlabel('(x1-mean1)/sqrt(M-1)')
plt.ylabel('(x2-mean2)/sqrt(M-1)')
```

## Now compute the SVD

```python
U, S, Vt = np.linalg.svd(Y,full_matrices=False)
print("Size X = ", X.shape)
print("Size Sigma = ",S.shape)
print("Size U = ",U.shape)
print("Size Vt = ",Vt.shape)
```

```python
print(Vt.T)
```

```python
variances = S**2
tot = sum(variances)
var_exp = [(i / tot)*100 for i in variances]
cum_var_exp = np.cumsum(var_exp)
print(var_exp)
```
