```python
import numpy as np
import numpy.linalg as la
import graph
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = np.array([[0,0,34],[3,0,0],[0,3,0]])
team_names = ['San Francisco','New York','Chicago']
```

```python
H1 = np.array([[0,0.48,0.98],[0.52,0,0.48],[0.02,0.52,0]])
print(H1)
```

```python
n = 3
H = np.zeros((n,n))
for i in range(n):
    col_sum = np.sum(H1[:,i])
    if col_sum != 0:
        H[:,i] = H1[:,i] / col_sum
    else:
        H[:,i] = 1/n
```

```python
graph.graph_matrix(H, team_names)
```

## Make a Google Matrix

```python
x0 = np.random.rand((n))
x0N=x0/la.norm(x0,1)

x = x0N.copy()
its = 20
allx = np.zeros((n,its))
allx[:,0] = x

for k in range(1,its):
    x = H.dot(x)
    allx[:,k] = x
    
plt.plot(allx.T)

print("x = ",x)
```
