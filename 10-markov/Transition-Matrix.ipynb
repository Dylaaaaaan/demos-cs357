## Make the transition matrix

```python
A  = np.array([
    [  0, 0.8, 1.0],
    [0.5,   0,   0],
    [0.5, 0.2,   0]
])
```

```python
graph.graph_matrix(A,['home', 'calendar', 'hw'])
```

### Try $x_1 \leftarrow A x_0$

```python
x0 = np.array([0.5, 0.5, 0.0])

x1 = A.dot(x0)
print(x1)
```

```python
print(la.norm(x1,1))
```

## Run the Power Method

Here we'll store all iterations in `allx` in order to view the running probabilities later.

```python
x = x0.copy()

its = 50

allx = np.zeros((3,its))
allx[:,0] = x

for k in range(1,its):
    x = A.dot(x)
    allx[:,k] = x
```

```python
plt.plot(allx.T)
plt.xlabel('')
```
