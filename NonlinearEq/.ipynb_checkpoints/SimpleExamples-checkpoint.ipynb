```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

# Spring problem

```python
def fSpring(x,k):
    return k*x

x = np.linspace(0, 0.1, 100)
# Spring constant
k = 40 #N/m
plt.plot(x, fSpring(x,k))
plt.xlabel('x [m]')
plt.ylabel('Spring force [N]')
plt.grid(True)
```

## Drag problem

```python
def fDrag(v,mu):
    return mu*v**2

v = np.linspace(0, 10, 100)
# Spring constant
mu = 0.5 #N.s/m
plt.plot(v, fDrag(v,mu))
plt.xlabel('v [m/s]')
plt.ylabel('Drag force [N]')
plt.grid(True)
```

```python
def fDragZero(v,mu,Fbar):
    return mu*v**2-Fbar


L = 0
R = 10
v = np.linspace(L, R, 100)
# Spring constant
mu = 0.5 #N.s/m
# Given force
F = 20


plt.plot(v, fDragZero(v,mu,F))
plt.xlabel('v [m/s]')
plt.ylabel('f(v) [N]')
plt.grid(True)
```

## Bisection method

```python
vtrue = np.sqrt(F/mu)
errors = []

a=L
b=R

for i in range(20):
    m = (a+b)/2

    if np.sign(fDragZero(a,mu,F)) == np.sign(fDragZero(m,mu,F)):
        a = m
    else:
        b = m

    errors.append(np.abs(vtrue - m))
    print("%10g %10g %10g" % (a, b, errors[-1]))
```

```python
for i in range(len(errors)-1):
    print(errors[i+1]/errors[i]**1.0)
```

```python
errors = np.array(errors)
plt.semilogy(errors[1:]/errors[:-1])
```
