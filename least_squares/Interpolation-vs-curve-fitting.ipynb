```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

## Using Least Square function

```python
# number of data points
m = 10
ti = np.linspace(0, 10, 10)
yi = np.sin(ti)+0.4*np.random.rand(10)*np.sin(ti)


def fitfunction(s,t,coeffs):
    f = 0*t
    for i in range(s):
        f = f + coeffs[i]*t**i    
    return f


# Least Squares Solution: data fitting using a cubic function
n = 3 + 1
A = np.array([
    ti**i
    for i in range(n)
    ]).T

coeffs,residual,rank,sval=np.linalg.lstsq(A,yi)
  

# Interpolation
V = np.array([
    ti**i
    for i in range(m)
]).T

coeffs2 = la.solve(V, yi)


plt.plot(ti, yi, 'o', label="Data points")
plt.plot(ti, fitfunction(n,ti,coeffs), 'm', label="Least-squares fit")
points = np.linspace(0, 10, 1000)
plt.plot(points, fitfunction(m,points,coeffs2), label="Interpolant")
#plt.plot(points, np.sin(points), label="sin")
plt.legend(loc="best")


```
