```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from math import factorial
```

```python
import sympy as sp
sp.init_printing()
x = sp.Symbol("x")
```

```python
f = sp.exp(x)

f
```

```python
def plot_sympy(my_f, my_pts, **kwargs):
    f_values = np.array([my_f.subs(x, pt) for pt in my_pts])
    plt.plot(pts, f_values, **kwargs)

    
def semilogy_sympy(my_f, my_pts, **kwargs):
    f_values = np.array([my_f.subs(x, pt) for pt in my_pts])
    plt.semilogy(pts, f_values, **kwargs)
```

```python
n = 4
xo = 2
```

```python
taylor = 0

for i in range(n):
    taylor += f.diff(x, i).subs(x, xo)/factorial(i) * (x-xo)**i

error =  f - taylor
```

```python
pts = np.linspace(-1, 4, 100)
plot_sympy(taylor, pts, label="taylor n=3")
plot_sympy(f, pts, label="f")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('function values')
```

```python
semilogy_sympy(error, pts, label="error")
semilogy_sympy(x**2, pts, label="$x$")
semilogy_sympy(x**3, pts, label="$x^3$")
semilogy_sympy(x**4, pts, label="$x^4$")
semilogy_sympy(x**5, pts, label="$x^5$")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('error')
plt.title('Error and bounds when n = 2')
```

```python
semilogy_sympy(error, pts, label="error")
semilogy_sympy((x-2), pts, label="$(x-2)$")
semilogy_sympy((x-2)**2, pts, label="$(x-2)**2$")
semilogy_sympy((x-2)**3, pts, label="$(x-2)**3$")
semilogy_sympy((x-2)**4, pts, label="$(x-2)**4$")
semilogy_sympy((x-2)**5, pts, label="$(x-2)**5$")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('error')
```
