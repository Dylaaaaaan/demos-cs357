```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from math import factorial
```

```python
import sympy as sp
sp.init_printing()
x = sp.Symbol("x")
```

```python
f = sp.exp(x)

f
```

```python
def plot_sympy(my_f, my_pts, **kwargs):
    f_values = np.array([my_f.subs(x, pt) for pt in my_pts])
    plt.plot(pts, f_values, **kwargs)

    
def semilogy_sympy(my_f, my_pts, **kwargs):
    f_values = np.array([my_f.subs(x, pt) for pt in my_pts])
    plt.semilogy(pts, f_values, **kwargs)
```

```python
n = 2
xo = 2
```

```python
taylor = 0

for i in range(n):
    taylor += f.diff(x, i).subs(x, xo)/factorial(i) * (x-xo)**i

error =  f - taylor
```

```python
pts = np.linspace(-1, 4, 100)
plot_sympy(taylor, pts, label="taylor n=1")
plot_sympy(f, pts, label="f")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('function values')
```

```python
Ox2_function = x**n
Oxminus2_function = (x-xo)**n
plot_sympy(error, pts, label="error")
plot_sympy(Ox2_function, pts, label="$x^n$")
plot_sympy(Oxminus2_function, pts, label="$(x-2)^n$")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('error')
plt.title('Error and bounds when n = 2')
```

```python
Ox2_function = x**n
Oxminus2_function = 10*(x-xo)**n
semilogy_sympy(error, pts, label="error")
semilogy_sympy(Ox2_function, pts, label="$x^n$")
semilogy_sympy(Oxminus2_function, pts, label="$(x-2)^n$")
plt.legend(loc="best")
plt.grid()
plt.xlabel('$x$')
plt.ylabel('error')
plt.title('Error and bounds when n = 2')
```
