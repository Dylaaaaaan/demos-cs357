```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt


import seaborn as sns
sns.set(font_scale=2)
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

## Take some data, scale to it's mean

```python
np.random.seed(8885)
Xhat = np.dot(np.random.rand(2, 2), np.random.randn(2, 200)).T
#Xhat = np.random.randn(2, 200).T
X = Xhat - Xhat.mean(axis=0)

plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
```

## Now compute the SVD

```python
U, S, V = np.linalg.svd(X)
print("Size X = ", X.shape)
print("Size Sigma = ",S.shape)
print("Size U = ",U.shape)
print("Size V = ",V.shape)
```
