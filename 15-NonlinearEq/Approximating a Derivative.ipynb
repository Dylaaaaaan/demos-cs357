```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(font_scale=2)
sns.set_style("whitegrid")
```

```python
def f(x):
    return np.exp(x) - 2

def df(x):
    return np.exp(x)
```

```python
x = np.linspace(-1, 2, 100)
plt.plot(x, f(x), lw=3)
```

```python
xx = 1.0
h = 1.0
errors = []
hs = []

for i in range(50):
    dfapprox = (f(xx+h) - f(xx)) / h
    
    err = np.abs(df(xx) - dfapprox)
    
    print(" %E \t %E " %(h, err) )
    
    hs.append(h)
    errors.append(err)
    h = h / 2
```

```python
plt.loglog(hs, errors, lw=3)
plt.xlabel('h')
plt.ylabel('error')
```
